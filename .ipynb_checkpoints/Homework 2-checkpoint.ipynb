```python
import matplotlib.pyplot as plt
import numpy as np
import typing 

#=======================================================
# Implementation of the function from Homework 2 using
# numpy functions for numerical analysis
# Parameters:
# - n : the index of the function, determines how many
#       terms are included in the expansion
# - x : a list of abscissa to compute f for
# Returns
# - a list of values f(x) for all provideed abscissa
#=======================================================
def f(n : int, x : np.ndarray[float]) -> np.ndarray[float]:
    
    total = np.pi/4                   # pi/4 gets multiplied to pi/2 later
    for i in range(1, n+1, 2):        # summing odd numbers 1 to n inclusive requires range(n+1)
        total += np.sin(x*i)/i        # for loop for sum
    return 2 * total                  # double everything, hence why pi/4 instead of pi/2

X = np.linspace(-np.pi, np.pi, 100)   # Make array of points between -pi and pi
f1 = f(11, X)                         # f(x) for n == 11
f2 = f(21, X)                         # f(x) for n == 21
f3 = f(31, X)                         # f(x) for n == 31

# Make plots
plt.plot(X, f1, label='n=11')
plt.plot(X, f2, label='n=21')
plt.plot(X, f3, label='n=31')
plt.legend()
plt.title("f_11(x), f_21(x), and f_31(x)")
```

```python
from sympy import *

#=======================================================
# Implementation of the function from Homework 2 using
# sympy functions for symbolic analysis
# Parameters:
# - n : the index of the function, determines how many
#       terms are included in the expansion
# - x : a list of abscissa to compute f for
# Returns
# - a list of values f(x) for all provideed abscissa
#=======================================================
def f(n : int, x : np.ndarray[float]) -> np.ndarray[float]:
    
    total = pi/4
    for i in range(1, n+1, 2):
        total += sin(x*i)/i
    
    return 2 * total   

# Integrate with respect to x
x = Symbol('x')
integrate(f(11, x), x)
```

```python
#=======================================================
# Implementation of the integral derived above using
# numpy functions for numerical analysis
# Parameters:
# - n : the index of the function, determines how many
#       terms are included in the expansion
# - x : a list of abscissa to compute f for
# Returns
# - a list of values f(x) for all provideed abscissa
#=======================================================
def g(n : int, x : np.ndarray[float]) -> np.ndarray[float]:
    
    total = np.pi*x/4
    for i in range(1, n+1, 2):
        total -= np.cos(x*i)/i**2
    return 2 * total


X = np.linspace(-np.pi, np.pi, 100)  # Make array of points between -pi and pi
g1 = g(11, X) - g(11, -np.pi)        # g(x) - g(-pi) with n == 11

# plotting
plt.plot(X, g1)
plt.title("Integral of f_11(x)")
```
